```python
import math
def radar_chart(data, categories):
    N = len(categories)
    values = data + data[:1]
    angles = [n / float(N) * 2 * math.pi for n in range(N)]
    angles += angles[:1]

    ax = plt.subplot(111, polar=True)
    plt.xticks(angles[:-1], categories)

    ax.plot(angles, values, linewidth=2, linestyle='solid')
    ax.fill(angles, values, alpha=0.4)

categories = ["Sales", "Quantity", "Discount", "Profit", "Shipping Cost"]
data = [df["Sales"].mean(), df["Quantity"].mean(), df["Discount"].mean(), df["Profit"].mean(), df["Shipping Cost"].mean()]
plt.figure(figsize=(8, 8))
radar_chart(data, categories)
plt.title("Radar Chart Example")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df["Sales"], df["Profit"], marker='o')
for i, txt in enumerate(df["Profit"]):
    plt.annotate(txt, (df["Sales"][i], df["Profit"][i]))
plt.title("Annotated Plot Example")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df["Sales"], df["Profit"], df["Quantity"], c='r', marker='o')
ax.set_xlabel("Sales")
ax.set_ylabel("Profit")
ax.set_zlabel("Quantity")
plt.title("3D Scatter Plot Example")
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Plot 1
axs[0, 0].plot(df["Sales"], label="Sales")
axs[0, 0].set_title("Sales Line Plot")
axs[0, 0].legend()

# Plot 2
axs[0, 1].bar(range(len(df)), df["Profit"], color='orange', label="Profit")
axs[0, 1].set_title("Profit Bar Plot")
axs[0, 1].legend()

# Plot 3
axs[1, 0].scatter(df["Quantity"], df["Shipping Cost"], color='green', label="Shipping Cost vs Quantity")
axs[1, 0].set_title("Scatter Plot")
axs[1, 0].legend()

# Plot 4
axs[1, 1].pie(df["Discount"], labels=range(len(df)), autopct='%1.1f%%', startangle=90)
axs[1, 1].set_title("Discount Pie Chart")

plt.tight_layout()
plt.show()
```
